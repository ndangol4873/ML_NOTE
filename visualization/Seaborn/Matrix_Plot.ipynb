```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

- axes level -> heatmap

### Heatmap

```python
# line plot
gap = px.data.gapminder()
gap
```

### Long vs wide data frame
- pivot helps to covert long into wide df

```python
temp_df = gap.pivot(index='country',columns='year', values='lifeExp')
temp_df
```

```python
## Heat Map
plt.figure(figsize=(15,15))
sns.heatmap(data=temp_df)
plt.show()
```

```python
europe_df = gap.loc[gap['continent']=='Europe'].pivot(index='country', columns='year', values='lifeExp')
europe_df
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(data=europe_df, annot=True, linewidths=.2, cmap='viridis')
plt.show()
```
