### Figure level function -> `catplot`

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# import datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

### Categorical Scatter Plot

- Stripplot
- Swarmplot


- Stripplot
- figure level -> Stripplot
- axes level -> catplot

```python
tips
```

```python
# strip plot
# axes level function
sns.stripplot(data=tips, x='day', y='total_bill')
```

```python
sns.stripplot(data=tips, x='day', y='total_bill',jitter=.01)
```

```python
# using catplot
# figure level function
sns.catplot(data=tips, x='day',y='total_bill', kind='strip', hue='sex')
```

- Swarmplot

```python
# using catplot
# axes level function
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
# using catplot
# figure level function

sns.catplot(data=tips, x='day',y='total_bill', kind='swarm', hue='sex')
plt.show()
```

### Box Plot
A boxplot is a standardized way of displaying the distribution of data based on a 
five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

- axis level function -> box plot -> voilion plot
- fig level function -> catplot

```python
tips.info()
```

```python
# Box plot
sns.boxplot(data=tips)
```

```python
sns.boxplot(data=tips, y='total_bill')
```

```python
sns.boxplot(data=tips, x='day', y='total_bill')
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker')
```

```python
sns.catplot(data=tips, x='day', y='total_bill',kind='box')
```

```python
sns.catplot(data=tips, x='day', y='total_bill',kind='box', hue='sex')
```

### Violinplot = (Boxplot + KDEplot)


```python
# violinplot
sns.violinplot(data=tips,x='day',y='total_bill')
plt.show()
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='smoker')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='smoker', split =True)
```

### Pointplot

```python
sns.pointplot(data=tips, x='sex', y='total_bill', hue='smoker',errorbar=None)
```

### countplot

A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative, variable

```python
sns.countplot(data=tips, x='sex',hue='day')
```

```python
# faceting using catplot
sns.catplot(data=tips, x='sex',y='total_bill',col='smoker',kind='box',row='time')
```
