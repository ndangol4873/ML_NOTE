### Surface Plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx,yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
### Some random fuction
z = np.sin(xx) + np.cos(yy)


fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(yy)


fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

### Contour Plots
- 2D representation of 3D plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2

fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2

fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
### Some random fuction
z = np.sin(xx) + np.cos(yy)


fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

### Heatmap

```python
delivery = pd.read_csv(f'{dataset}/IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
temp_df = delivery.loc[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
temp_df
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
grid
```

```python
plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.colorbar()
```
