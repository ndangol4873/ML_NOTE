### Specify Graph color

```python
plt.plot(batsman['index'],batsman['V Kohli'], color = 'green')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'red')
plt.title("Virat Kohli vs Rohit Sharma Career Comparision")
plt.xlabel("Season")
plt.ylabel("Run Scored")
```

### Line Style

```python
# plt.plot(batsman['index'],batsman['V Kohli'], color = 'green', linestyle = 'dashed')
plt.plot(batsman['index'],batsman['V Kohli'], color = 'green', linestyle = 'dashdot')

plt.plot(batsman['index'],batsman['RG Sharma'],color = 'red',  linestyle = 'dotted')
plt.title("Virat Kohli vs Rohit Sharma Career Comparision")
plt.xlabel("Season")
plt.ylabel("Run Scored")
```

### Line Style width

```python
# plt.plot(batsman['index'],batsman['V Kohli'], color = 'green', linestyle = 'dashed')
plt.plot(batsman['index'],batsman['V Kohli'], color = 'green', linestyle = 'solid', linewidth = 2)

plt.plot(batsman['index'],batsman['RG Sharma'],color = 'red',  linestyle = 'dotted', linewidth = 4)
plt.title("Virat Kohli vs Rohit Sharma Career Comparision")
plt.xlabel("Season")
plt.ylabel("Run Scored")
```

### Markers

```python
# plt.plot(batsman['index'],batsman['V Kohli'], color = 'green', linestyle = 'dashed')
plt.plot(batsman['index'],batsman['V Kohli'], color = 'green', linestyle = 'solid', linewidth = 4, marker='D',markersize = 5)

plt.plot(batsman['index'],batsman['RG Sharma'],color = 'red',  linestyle = 'dotted', linewidth = 2,  marker='*', markersize = 10)
plt.title("Virat Kohli vs Rohit Sharma Career Comparision")
plt.xlabel("Season")
plt.ylabel("Run Scored")
```

### Label

```python
# plt.plot(batsman['index'],batsman['V Kohli'], color = 'green', linestyle = 'dashed')
plt.plot(batsman['index'],batsman['V Kohli'], color = 'green', linestyle = 'solid', linewidth = 4, marker='D',markersize = 5, label = "Virat")

plt.plot(batsman['index'],batsman['RG Sharma'],color = 'red',  linestyle = 'dotted', linewidth = 2,  marker='*', markersize = 10, label = "Rohit")
plt.title("Virat Kohli vs Rohit Sharma Career Comparision")
plt.xlabel("Season")
plt.ylabel("Run Scored")
plt.legend(loc='best') ## Default Parameter
```

### Grids option

```python
# plt.plot(batsman['index'],batsman['V Kohli'], color = 'green', linestyle = 'dashed')
plt.plot(batsman['index'],batsman['V Kohli'], color = 'green', linestyle = 'solid', linewidth = 4, marker='D',markersize = 5, label = "Virat")

plt.plot(batsman['index'],batsman['RG Sharma'],color = 'red',  linestyle = 'dotted', linewidth = 2,  marker='*', markersize = 10, label = "Rohit")
plt.title("Virat Kohli vs Rohit Sharma Career Comparision")
plt.xlabel("Season")
plt.ylabel("Run Scored")
plt.legend(loc='best') ## Default Parameter
plt.grid()
```

### Trim

```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
```

```python
# Adding Outlier 450000 in 2021
price = [48000,54000,57000,49000,47000,45000, 450000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
```

```python
# Adding Outlier 
price = [48000,54000,57000,49000,47000,45000, 450000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0, 75000)
```
