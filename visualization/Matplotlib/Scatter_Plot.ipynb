### Scatter Plot

```python
# plt.scatter simple function
x = np.linspace(-10,10,50)

y = 10*x + 3 + np.random.randint(0,300,50)
y
```

```python
plt.scatter(x,y)
```

```python
# plt.scatter on pandas data
df = pd.read_csv(f"{dataset}/batter.csv")
df = df.head(200)
df
```

```python
plt.scatter(df['avg'], df['strike_rate'], color = 'red', marker="*")
plt.title("Avg & SR Analysis of the Batsmen")
plt.ylabel("Strike_Rate")
plt.xlabel("Average Rate")
```

### Marker

```python
tips = sns.load_dataset('tips')
tips
```

```python
plt.scatter(tips['total_bill'], tips['tip'])
```

```python
# scatterplot using plt.Scatter
# slower
plt.scatter(tips['total_bill'], tips['tip'],s=tips['size']*10)
```

```python
# scatterplot using plt.plot
# faster
plt.plot(tips['total_bill'], tips['tip'],'o')
```
