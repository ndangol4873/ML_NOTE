## Color and label

```python
# color and label
df = pd.read_csv(f'{dataset}/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'], df['2015'])
plt.bar(df['batsman'], df['2016'])
```

### Multiple Bars

```python
x = np.arange(df.shape[0])
print(x)
```

```python
plt.bar(x, df['2015'],width=0.2, color='red', label ='2015')
plt.bar(x + .2, df['2016'],width=0.2, label ='2016')
plt.bar(x + .4, df['2017'],width=0.2, label ='2017')

plt.legend()
plt.xticks(x, df['batsman'])

```

### Plt.show()

```python
size = 0.2
plt.bar(x, df['2015'],width=size, color='red', label ='2015')
plt.bar(x + size, df['2016'],width=size, label ='2016')
plt.bar(x + (size*2), df['2017'],width=size, label ='2017')

plt.xlabel("Batsmen")
plt.ylabel("Max Run Scored")
plt.legend()
plt.xticks(x, df['batsman'])
plt.show()

```

### Stagged Bar Chart

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2017'] + df['2016']),label='2015')

plt.legend()
plt.show()
```
