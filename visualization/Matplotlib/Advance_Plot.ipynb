```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import os
from pathlib import Path 
dataset =Path("D:\ML_NOTE_DATASET")
```

## Colored Scatter Plot

```python
iris = pd.read_csv(f'{dataset}\iris.csv')
iris.head()
```

```python
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

```python
## Replacing the Categorical value
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

```python
## Color Bar
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='jet')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
plt.show()
```

```python
## Opacity
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='jet',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
plt.show()
```

## Plot Size

```python
## Size
plt.figure(figsize=(10,10))

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='jet',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
plt.show()
```

```python
## Applying in the Data set
batters = pd.read_csv(f'{dataset}/batter.csv')
batters.head()
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
sample_df
```

```python
plt.figure(figsize=(18,12))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs']/25)
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])

plt.show()
```

## Horizontal and Vertical lines

```python
plt.figure(figsize=(18,12))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs']/25)

plt.axhline(130,color='red') ## Plotting Horizontal Lines
plt.axhline(140,color='green') ## Plotting Horizontal Lines
plt.axvline(30,color='blue') ## Plotting Vertical Lines

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])

plt.show()
```

#### Multiple Plots

```python
fig,ax = plt.subplots(nrows=3, ncols=1, figsize=(10,10),sharex=True)

ax[0].scatter(batters['avg'],batters['strike_rate'],color='green')
ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].scatter(batters['avg'],batters['runs'])
ax[1].set_title('Avg Vs Run Rate')
ax[1].set_ylabel('Runs')


ax[2].scatter(batters['avg'],batters['strike_rate'],color = 'red',alpha= 0.1)
ax[2].set_title('Avg Vs Strike Rate')
ax[2].set_ylabel('Strike Rate')
ax[2].set_xlabel('AVG')

fig.show()
```

## Scatter Plot and Histogram

```python
fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(10,10))

## Scatter Plot
ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='green')
ax[0,0].set_title('Avg Vs Strike Rate')
ax[0,0].set_ylabel('Strike Rate')

ax[0,1].scatter(batters['avg'],batters['runs'])
ax[0,1].set_title('Avg Vs Run Rate')
ax[0,1].set_ylabel('Runs')

##Histogram
ax[1,0].hist(batters['avg'])
ax[1,0].set_title('Avg Rate Hist')
ax[1,0].set_ylabel('Runs hist')



ax[1,1].hist(batters['strike_rate'])
ax[1,1].set_title('Strike Rate Hist')
ax[1,1].set_ylabel('Runs hist2')
fig.show()
```
